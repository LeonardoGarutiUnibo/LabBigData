```python
matrimoni_data_grouped = matrimoni_data_cleaned.groupby(["TIME", "Rito"])["Value"].sum().reset_index()

sns.set_theme(style="whitegrid")

plt.figure(figsize=(12, 6))
sns.lineplot(data=matrimoni_data_grouped, x="TIME", y="Value", hue="Rito", marker="o")

plt.title("Andamento dei matrimoni per tipo di rito (2009-2019)", fontsize=16)
plt.xlabel("Anno", fontsize=12)
plt.ylabel("Numero di matrimoni", fontsize=12)
plt.legend(title="Tipo di Rito", fontsize=10)
plt.xticks(matrimoni_data_grouped["TIME"].unique(), rotation=45)
plt.tight_layout()

plt.show()
```
